```python
def assess_risk(weather, speed, traffic_density):
    risk_score = 0
    if weather.lower() in ['rainy', 'foggy', 'snowy']:
        risk_score += 3
    if speed > 100:
        risk_score += 3
    elif speed > 80:
        risk_score += 2
    else:
        risk_score += 1
    if traffic_density == 'high':
        risk_score += 3
    elif traffic_density == 'medium':
        risk_score += 2
    else:
        risk_score += 1
    return 'High Risk' if risk_score >= 7 else 'Moderate Risk' if risk_score >= 5 else 'Low Risk'

# Sample test
print("Accident Risk:", assess_risk('Rainy', 90, 'High'))
```
